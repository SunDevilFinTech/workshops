```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_data = np.random.random(50) * 100
Y_data = np.random.random(50) * 100
```

```python
# scatter plot

# c is color, marker is point pattern, s is size, alpha is opacity
plt.scatter(X_data, Y_data, c="red", marker="*", s=100, alpha=0.3) # everything after data is optional
plt.show()
```

```python
# line plot

years = [2006 + x for x in range(16)]
weights = np.random.randint(78, 85, size=16)

# lw = line width
plt.plot(years, weights, c="r", lw=3, linestyle="--")
```

```python
# Bar plots

x = ["C++", "C#", "Java", "Python", "Go"]
y = [20, 15, 10, 100, 55]

# color not c as parameter
plt.bar(x, y, color="g", align="edge", width=0.5, edgecolor="r", lw=6)
plt.show()
```

```python
# Histogram

ages = np.random.normal(20, 1.5, 1000)

plt.hist(ages, bins=10)
plt.show()

plt.hist(ages, bins=[ages.min(), 18, 21, ages.max()])
plt.show()

plt.hist(ages, cumulative=True)
plt.show()
```

```python
# Pie Charts

x = ["C++", "C#", "Java", "Python", "Go"]
y = [20, 15, 10, 100, 55]
# pull out Java a little
explodes = [0, 0, 0.3, 0, 0]

plt.pie(y, labels=x, explode=explodes)
plt.show()

plt.pie(y, labels=x, explode=explodes, autopct="%.2f%%", pctdistance=1.2, startangle=90)
plt.show()
```

```python
# boxplot

heights = np.random.normal(172, 8, 300)

plt.boxplot(heights)
plt.show()
```

```python
income = [55, 56, 62, 61, 72, 72, 74, 75]
years = [2014 + i for i in range(8)]

income_tickers = list(range(50, 81, 2))

plt.plot(years, income)
plt.title("Income in USD")
plt.xlabel("Year")
plt.ylabel(income_tickers)
plt.show()
```
