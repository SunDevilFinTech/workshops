```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a NumPy array
array = np.array([1, 2, 3, 4, 5]) # array is like a list
print("NumPy Array:", array)
```

```python
# Array operations
array_squared = array ** 2
print("Squared Array:", array_squared)

# Calculate the mean
mean_value = np.mean(array)
print("Mean Value:", mean_value)

# Calculate the standard deviation
std_dev = np.std(array)
print("Standard Deviation Value:", std_dev)
```

```python
import pandas as pd

# Create a DataFrame for transactions - think of dataframe like an excel sheet in code
transactions_df = pd.DataFrame({
    'Transaction': [100, -20, 50, -30],
    'Type': ['Deposit', 'Withdraw', 'Deposit', 'Withdraw']
})
print("Transactions DataFrame:")
print(transactions_df)

# Calculate total balance
total_balance = transactions_df['Transaction'].sum()
print("Total Balance:", total_balance)
```

```python
import matplotlib.pyplot as plt

# Plotting transactions
plt.figure(figsize=(8, 5))
plt.bar(transactions_df.index, transactions_df['Transaction'], color=['green' if x > 0 else 'red' for x in transactions_df['Transaction']])
plt.title('Transaction Amounts')
plt.xlabel('Transaction Index')
plt.ylabel('Amount')
plt.axhline(0, color='black', linewidth=0.8)  # Add a line at y=0
plt.show()
```

```python
# Install yfinance (uncomment if needed)
# !pip install yfinance

import yfinance as yf

# Fetch data for a specific stock (e.g., Apple)
ticker = "AAPL"
data = yf.download(ticker, start="2023-01-01", end="2023-12-31")
print("Apple Stock Data:")
print(data.head())

# Plot closing prices
plt.figure(figsize=(10, 6))
plt.plot(data['Close'], label='AAPL Closing Price', color='blue')
plt.title('Apple Stock Closing Prices in 2023')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```
